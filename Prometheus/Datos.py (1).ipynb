```python
from prometheus_client import start_http_server, Gauge
import datetime
import random
import time

# --- CONFIGURACIÓN ---
PROVINCIAS = ["Santo Domingo", "La Sábana", "Paraíso"]

RANGOS_TEMPERATURA = {
    "Santo Domingo": (22.0, 27.0),
    "La Sábana": (23.0, 28.0),
    "Paraíso": (19.0, 26.0)
}

# Creamos una métrica tipo Gauge (valor numérico que puede subir o bajar)
temperatura_gauge = Gauge(
    'temperatura_ciudad_celsius',  # nombre de la métrica
    'Temperatura simulada por provincia',  # descripción
    ['ciudad']  # etiqueta (label)
)

def generar_y_actualizar_metricas():
    """
    Genera valores aleatorios de temperatura y los actualiza en las métricas.
    """
    for provincia in PROVINCIAS:
        min_temp, max_temp = RANGOS_TEMPERATURA[provincia]
        temperatura_aleatoria = random.uniform(min_temp, max_temp)
        temperatura_gauge.labels(ciudad=provincia).set(temperatura_aleatoria)
    print(f"[{datetime.datetime.now()}] Métricas actualizadas.")

if __name__ == '__main__':
    # Inicia un servidor HTTP en el puerto 8000
    start_http_server(8000)
    print("Servidor de métricas iniciado en http://localhost:8000/metrics")

    # Bucle infinito para actualizar cada 15 segundos
    while True:
        generar_y_actualizar_metricas()
        time.sleep(15)
```
